```python
import pandas as pd 
```

```python
social_df = pd.read_csv('social.csv')
social_df.head(10)
```

```python
social_df.dtypes
```

```python
social_df["TIME"]=social_df["TIME"].astype(str)
```

```python
social_df["LocationTime"] = social_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=social_df
social_df.head(10)
```

```python
social_df["TIME"]=social_df["TIME"].astype(int)
```

```python
social_df.dtypes
```

```python
social_df = social_df.drop(columns=['Flag Codes', 'MEASURE', 'SUBJECT', 'FREQUENCY'])
social_df.head(10)
```

```python
social_df.isnull().any()
```

```python
social_time_df = social_df.loc[(social_df['TIME'] >= 1990) & (social_df['TIME'] <= 2020)]
social_time_df
```

```python
social_clean_df = social_time_df[['LocationTime','LOCATION','INDICATOR','TIME', 'Value']]
social_clean_df
```

```python
social_clean_df.to_csv("social_support_clean.csv", index=False)
```
