```python
import pandas as pd
```

```python
#import the data
employment_df= pd.read_csv("Employment_1990_2020.csv")
employment_df.head(10)
```

```python
#Change datatype to be accurate for time
employment_df["TIME"]=employment_df["TIME"].astype(str)
employment_df.dtypes
```

```python
#Combine Location and Time into Single Column
employment_df["LocationTime"] = employment_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=employment_df
employment_df.head(10)
```

```python
# Change the Time Column dataformat back to int
employment_df["TIME"]=employment_df["TIME"].astype(int)
employment_df.dtypes
```

```python
# Drop null values
employment_df.isnull().any()
```

```python
#Drop Columns
employment_df = employment_df.drop(columns=['employment_val_BUPPSRY_PC_25_64','employment_val_UPPSRY_NTRY_PC_25_64'])
employment_df.head(10)
```

```python
#Clean up time column
#employment_df["TIME"]= employment_df["TIME"].str.slice(0,4)
```

```python
# Total number of rows
num_rows = len(employment_df)
print(num_rows)
```

```python
# TImeframe 1990-2020
employment_df= employment_df.loc[(employment_df['TIME'] >= 1990) & (employment_df['TIME'] <= 2020)]
```

```python
#Change datatype to be accurate for time
employment_df["TIME"]=employment_df["TIME"].astype(int)
employment_df.dtypes
```

```python
# Rename value column
employment_df=employment_df.rename(columns={"employment_val_TRY_PC_25_64":"Emp_val"})
employment_df.head(10)
```
