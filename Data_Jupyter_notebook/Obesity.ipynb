```python
import pandas as pd
```

```python
#import the data
obesity_df= pd.read_csv("../Resources/Obesity_1990_2020.csv")
```

```python
#view top 10 
obesity_df.head(10)
```

```python
#change Datatype of time to Str
obesity_df["TIME"]=obesity_df["TIME"].astype(str)
```

```python
#Combine Location and Time into Single Column
obesity_df["LocationTime"] = obesity_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=obesity_df
```

```python
#change Datatype of time back to int
obesity_df["TIME"]=obesity_df["TIME"].astype(int)
```

```python
obesity_df
```

```python
#filter to only include years 1990-2020
obesity_df_filter= obesity_df.loc[(obesity_df['TIME'] >= 1990) & (obesity_df['TIME'] <= 2020)]
obesity_df_filter
```

```python
#drop Extra columns
obesity_df_clean=obesity_df_filter.drop(columns=["Flag Codes", "MEASURE", "FREQUENCY"])
obesity_df_clean
```

```python
obesity_df_clean = obesity_df_clean[['LocationTime','LOCATION','INDICATOR','SUBJECT','TIME', 'Value']]
obesity_df_clean
```

```python
obesity_df_clean.to_csv("obesity_clean.csv", index=False)
```
