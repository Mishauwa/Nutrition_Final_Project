```python
import pandas as pd
```

```python
df = pd.read_csv('Work_Hour.csv') 
df
```

```python
df["TIME"]=df["TIME"].astype(str)
```

```python
df["LocationTime"] = df[["LOCATION", "TIME"]].apply('-'.join, axis=1)
```

```python
df
```

```python
df = df.drop(columns=['Flag Codes'])
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df = df.drop(columns=['MEASURE'])
```

```python
df = df.drop(columns=['FREQUENCY'])
```

```python
df = df.drop(columns=['INDICATOR'])
```

```python
df
```

```python
df['TIME']=df['TIME'].astype(int)
```

```python
df = df.loc[(df['TIME'] >= 1990) & (df['TIME'] <= 2020)]
```

```python
df = df[['LocationTime', 'LOCATION', 'SUBJECT', 'TIME', 'Value']]
df
```

```python
df=df.rename(columns={"Value": "Work_Hour_Value"})
```

```python
df['Poverty_Value']=df['Work_Hour_Value'].astype(str)
```

```python
df=pd.pivot(df, index = ["LocationTime", "LOCATION", "TIME"], columns=["SUBJECT"], values=["Work_Hour_Value"])
```

```python
df.columns = ['_'.join(str(s).strip() for s in col if s) for col in df.columns] 
df.reset_index(inplace=True) 
df
```

```python
df["TIME"]=df["TIME"].astype(str) 
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.to_csv('Work_Hour_Cleaned.csv', index = False)
```
