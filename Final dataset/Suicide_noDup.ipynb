```python
import pandas as pd
import os
```

```python
# IMPORTING CSV FILE-OECD
oecd_data = os.path.join('Suicide_OECD_Clean.csv')
oecd_df = pd.read_csv(oecd_data)
oecd_df.head()
```

```python
#Drop Columns
oecd_df = oecd_df.drop(columns=['Suicide_Val_MEN','Suicide_Val_WOMEN'])
display(oecd_df)
```

```python
#Drop Columns
oecd_df = oecd_df.drop(columns=['LOCATION','TIME'])
oecd_df.head()
```

```python
# IMPORTING CSV FILE-Alcohol
alc_data = os.path.join('Alcohol_Clean.csv')
alc_df = pd.read_csv(alc_data)
alc_df.head()
```

```python
#Drop Columns
alc_df = alc_df.drop(columns=['LOCATION','TIME'])
alc_df.head()
```

```python
# Merge OECD and Alcohol
df1 = pd.merge(oecd_df, alc_df, on='LocationTime', how='left')
display(df1)
```
