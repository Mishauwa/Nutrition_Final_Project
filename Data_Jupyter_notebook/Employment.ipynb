```python
import pandas as pd
```

```python
#import the data
employment_df= pd.read_csv("../Resources/Employment_1990_2020.csv")
```

```python
#view top 10 
employment_df.head(10)
```

```python
print(employment_df["MEASURE"].unique())
```

```python
#Combine Location and Time into Single Column
employment_df["LocationTime"] = employment_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=employment_df
employment_df.head(10)
```

```python
#Clean up time column
employment_df["TIME"]= employment_df["TIME"].str.slice(0,4)
```

```python
#check Data types
employment_df.dtypes
```

```python
#Change datatype to be accurate for time
employment_df["TIME"]=employment_df["TIME"].astype(int)
employment_df.dtypes
```

```python
#filter to only includ 1990-2020
#remove extra columns not needed - Michael
employment_df= employment_df.loc[(employment_df['TIME'] >= 1990) & (employment_df['TIME'] <= 2020)]
#remove extra columns not needed
employment_df=employment_df.drop(columns=["Flag Codes","FREQUENCY", 'INDICATOR']).rename(columns={"Value": "employment_val"})
employment_df
```

```python
employment_df=pd.pivot_table(employment_df, index = ["LocationTime", "LOCATION", "TIME"], columns=["SUBJECT", "MEASURE"], values=["employment_val"])
employment_df
```

```python
#format column names
employment_df.columns = ['_'.join(str(s).strip() for s in col if s) for col in employment_df.columns]

#reset index
employment_df.reset_index(inplace=True)
```

```python
employment_clean=employment_df
employment_clean
```

```python
employment_clean.to_csv("employment_clean.csv", index=False)
```
