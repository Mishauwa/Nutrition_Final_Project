```python
import pandas as pd 
```

```python
df = pd.read_csv('Poverty.csv') 
df.head(10)
```

```python
df.dtypes
```

```python
df["TIME"]=df["TIME"].astype(str)
```

```python
df["LocationTime"] = df[["LOCATION", "TIME"]].apply('-'.join, axis=1)
```

```python
df
```

```python
df = df.drop(columns=['Flag Codes'])
```

```python
df['TIME'] = df["TIME"].astype(int)
```

```python
df
```

```python
df.isnull()
```

```python
df.dropna()
```

```python
df = df.drop(columns=['MEASURE'])
```

```python
df = df.drop(columns=['FREQUENCY'])
```

```python
df
```

```python
df = df.loc[(df['TIME'] >= 1990) & (df['TIME'] <= 2020)]
```

```python
df
```

```python
df = df[['LocationTime', 'LOCATION', 'INDICATOR', 'SUBJECT', 'TIME', 'Value']]
df
```

```python
df.to_csv('Poverty_Cleaned', index = False)
```
