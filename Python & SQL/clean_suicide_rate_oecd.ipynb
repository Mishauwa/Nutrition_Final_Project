```python
import pandas as pd
import os
```

```python
suicide_df = pd.read_csv('suicide_rate_oecd.csv')
suicide_df.head(10)
```

```python
suicide_df.dtypes
```

```python
suicide_df["TIME"]=suicide_df["TIME"].astype(str)
```

```python
suicide_df["LocationTime"] = suicide_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=suicide_df
suicide_df.head(10)
```

```python
suicide_df["TIME"]=suicide_df["TIME"].astype(int)
```

```python
suicide_df.dtypes
```

```python
suicide_df = suicide_df.drop(columns=['Flag Codes', 'MEASURE', 'INDICATOR','FREQUENCY'])
suicide_df.head(10)
```

```python
suicide_df.isnull().any()
```

```python
suicide_time_df = suicide_df.loc[(suicide_df['TIME'] >= 1990) & (suicide_df['TIME'] <= 2020)]
suicide_time_df
```

```python
suicide_clean_df = suicide_time_df[['LocationTime','LOCATION','SUBJECT','TIME','Value']]
suicide_clean_df
```

```python
suicide_rename_df = suicide_clean_df.rename(columns={'Value':'Suicide_val'})
suicide_rename_df
```

```python
suicide_oecd_df= suicide_rename_df.loc[(suicide_rename_df['SUBJECT'] == 'WOMEN') | (suicide_rename_df['SUBJECT'] == 'MEN')]
suicide_oecd_df
```

```python
print(suicide_rename_df['SUBJECT'].unique())
```

```python
ss_df=pd.pivot(suicide_rename_df, index = ["LocationTime", "LOCATION", "TIME"], columns=["SUBJECT"], values=["Suicide_val"])
ss_df
```

```python
#format column names
ss_df.columns = ['_'.join(str(s).strip() for s in col if s) for col in ss_df.columns]
#reset index
ss_df.reset_index(inplace=True)
```

```python
ss_df
```

```python
ss_df.isnull().sum()
```

```python
ss_df.to_csv("suicide_oecd_clean.csv", index=False)
```
