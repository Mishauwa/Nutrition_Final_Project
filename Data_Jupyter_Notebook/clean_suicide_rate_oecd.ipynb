```python
import pandas as pd 
```

```python
suicide_df = pd.read_csv('suicide_rate_oecd.csv')
suicide_df.head(10)
```

```python
suicide_df.dtypes
```

```python
suicide_df["TIME"]=suicide_df["TIME"].astype(str)
```

```python
suicide_df["LocationTime"] = suicide_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=suicide_df
suicide_df.head(10)
```

```python
suicide_df["TIME"]=suicide_df["TIME"].astype(int)
```

```python
suicide_df.dtypes
```

```python
suicide_df = suicide_df.drop(columns=['Flag Codes', 'MEASURE', 'FREQUENCY'])
suicide_df.head(10)
```

```python
suicide_df.isnull().any()
```

```python
suicide_time_df = suicide_df.loc[(suicide_df['TIME'] >= 1990) & (suicide_df['TIME'] <= 2020)]
suicide_time_df
```

```python
suicide_clean_df = suicide_time_df[['LocationTime','LOCATION','INDICATOR','SUBJECT','TIME', 'Value']]
suicide_clean_df
```

```python
suicide_clean_df.to_csv("suicide_oecd_clean.csv", index=False)
```
