```python
# import dependencies
import pandas as pd
```

```python
# import the data
suicide_rate_df = pd.read_csv('Suicide_Rates.csv')
# View Data
suicide_rate_df.head(10)
```

```python
# Data Types
suicide_rate_df.dtypes
```

```python
# change countries and year data types to string
suicide_rate_df['country'] = suicide_rate_df['country'].astype(str)
suicide_rate_df['year'] = suicide_rate_df['year'].astype(str)

suicide_rate_df.dtypes
```

```python
suicide_rate_df = suicide_rate_df.drop("HDI for year", axis=1)
suicide_rate_df = suicide_rate_df.drop("generation", axis=1)
```

```python
suicide_rate_df.dtypes
```

```python
# Remove 'years' suffixes by replacing with and empty string
suicide_rate_df['age'] = suicide_rate_df['age'].str.replace('years', '')
```

```python
suicide_rate_df.head(10)
```

```python
suicide_clean_df = suicide_rate_df
suicide_clean_df
```

```python
suicide_clean_df.to_csv("suicide_clean.csv", index=False)
```
