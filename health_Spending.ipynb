```python
import pandas as pd
```

```python
# import the data
health_spending_df = pd.read_csv("Healthy_spending_1990-2020.csv")
```

```python
# view head
health_spending_df.head(10)
```

```python
# Data types
health_spending_df.dtypes
```

```python
# change dataframe 
health_spending_df["TIME"]=health_spending_df["TIME"].astype(str)
```

```python
health_spending_df["LocationTime"] = health_spending_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=health_spending_df
health_spending_df.head()
```

```python
health_spending_df["TIME"]=health_spending_df["TIME"].astype(int)
```

```python
# check data types 
health_spending_df.dtypes
```

```python
health_spending_df = health_spending_df.drop(columns=['Flag Codes', 'MEASURE', 'SUBJECT', 'FREQUENCY'])
health_spending_df.head(10)
```

```python
health_spending_df.isnull().any()
```

```python
health_spending_df = health_spending_df.loc[(health_spending_df['TIME'] >= 1990) & (health_spending_df['TIME'] <= 2020)]
health_spending_df.head()
```

```python
health_clean_df = health_spending_df[['LocationTime', 'LOCATION', 'INDICATOR', 'TIME', 'Value']]
health_clean_df
```

```python
health_clean_df.to_csv("health_spending_clean.csv", index=False)
```
