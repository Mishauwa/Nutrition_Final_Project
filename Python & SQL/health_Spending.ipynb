```python
import pandas as pd
```

```python
# import the data
health_spending_df = pd.read_csv("Healthy_spending_1990-2020.csv")
```

```python
# view head
health_spending_df.head(10)
```

```python
# Data types
health_spending_df.dtypes
```

```python
# change dataframe 
health_spending_df["TIME"]=health_spending_df["TIME"].astype(str)
```

```python
health_spending_df["LocationTime"] = health_spending_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=health_spending_df
health_spending_df.head()
```

```python
health_spending_df["TIME"]=health_spending_df["TIME"].astype(int)
```

```python
# check data types 
health_spending_df.dtypes
```

```python
health_spending_df = health_spending_df.drop(columns=['Flag Codes', 'MEASURE', 'FREQUENCY', 'INDICATOR'])
health_spending_df.head(10)
```

```python
health_spending_df.isnull().any()
```

```python
health_spending_df = health_spending_df.loc[(health_spending_df['TIME'] >= 1990) & (health_spending_df['TIME'] <= 2020)]
health_spending_df.head()
```

```python
# remove year in locationTime
health_spending_df ['LocationTime'] = health_spending_df['LocationTime'].replace('-    ', '')
```

```python
# rename value to hSpend_val
health_spending_df = health_spending_df.rename(columns={"Value": "hSpend_val"})
health_spending_df.head(10)
```

```python
# rearrange the cols
health_clean_df = health_spending_df[['LocationTime', 'LOCATION','TIME', 'SUBJECT', 'hSpend_val']]
#health_clean_df.head(10)
```

```python
health_clean_df.head()
```

```python
# use pivot table to get the subject col values
health_spending_df = pd.pivot_table(health_spending_df, index =["LocationTime", 'LOCATION', 'TIME',],  columns=['SUBJECT'], values=["hSpend_val"] )
health_spending_df
```

```python
#format column names
health_spending_df.columns = ['_'.join(str(s).strip() for s in col if s) for col in health_spending_df.columns]
#reset index
health_spending_df.reset_index(inplace=True)
health_spending_df
```

```python
health_spending_df.head(10)
```

```python
# rearrange the cols
health_clean_df = health_spending_df
health_clean_df
```

```python
health_clean_df = health_clean_df.drop(columns=['hSpend_val_OOPEXP', 'hSpend_val_TOT'])
health_clean_df.head(10)
```

```python
health_clean_df.isnull().sum()
```

```python
health_clean_df=health_clean_df.drop(columns=["hSpend_val_COMPULSORY"]).rename(columns={"hSpend_val_VOLUNTARY":"Healthy_Spend_Val"})
health_clean_df
```

```python
health_clean_df.to_csv("healthy_spending_clean.csv", index=False)
```
