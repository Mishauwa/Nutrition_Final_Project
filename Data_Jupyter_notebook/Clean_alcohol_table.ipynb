```python
import pandas as pd 
```

```python
alcohol_df = pd.read_csv('alcohol.csv')
alcohol_df.head(10)
```

```python
alcohol_df.dtypes
```

```python
alcohol_df["TIME"]=alcohol_df["TIME"].astype(str)
```

```python
alcohol_df["LocationTime"] = alcohol_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=alcohol_df
alcohol_df.head(10)
```

```python
alcohol_df["TIME"]=alcohol_df["TIME"].astype(int)
```

```python
alcohol_df.dtypes
```

```python
alcohol_df = alcohol_df.drop(columns=['Flag Codes', 'MEASURE', 'SUBJECT', 'FREQUENCY', 'INDICATOR'])
alcohol_df.head(10)
```

```python
alcohol_df.isnull().any()
```

```python
 # alcohol_df["TIME"].loc['1990':'2020']
alcohol_time_df = alcohol_df.loc[(alcohol_df['TIME'] >= 1990) & (alcohol_df['TIME'] <= 2020)]
alcohol_time_df
```

```python
alcohol_clean_df = alcohol_time_df[['LocationTime','LOCATION','TIME', 'Value']]
alcohol_clean_df
```

```python
alcohol_rename_df = alcohol_clean_df.rename(columns={'Value':'Alcohol_val'})
alcohol_rename_df
```

```python
alcohol_rename_df.to_csv("alcohol_clean.csv", index=False)
```
