```python
#Import libraries and read data csv file 
import pandas as pd
education_df = pd.read_csv("../Final Project/Resources/Education_1990_2020.csv")
education_df.head(10)
```

```python
# Check the data types
education_df.dtypes
```

```python
# Converting Time int to string
education_df["TIME"]= education_df["TIME"].astype(str)
```

```python
#Combine Location and Time into Single Column
education_df["LocationTime"] = education_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame = education_df
education_df.head(10)
```

```python
# Change the Time Column dataformat back to int
education_df["TIME"]=education_df["TIME"].astype(int)
education_df.dtypes
```

```python
# Drop null values
education_df.isnull().any()
```

```python
#Drop Columns
education_df = education_df.drop(columns=['Flag Codes','FREQUENCY','MEASURE'])
education_df.head(10)
```

```python
# Total number of rows
num_rows = len(education_df)
print(num_rows)
```

```python
# Filter TIME values between 1990 and 2020
edu_filtered_df= education_df.loc[(education_df['TIME'] >= 1990) & (education_df['TIME'] <= 2020)]
edu_filtered_df.head(10)
```

```python
# Move LocationTime row to the beginning of the DataFrame
move_column = edu_filtered_df.pop('LocationTime')
edu_filtered_df.insert(0, 'LocationTime', move_column)
edu_filtered_df.head(10)
```

```python
# Total number of rows
num_rows = len(edu_filtered_df)
print(num_rows)
```

```python
# Save in a new csv file
edu_filtered_df.to_csv('Education_data.csv', index=False)
```
