```python
# import dependencies
import pandas as pd
from pycountry import countries
from pycountry_convert import (country_name_to_country_alpha3)
```

```python
# import the data
filtered_df = pd.read_csv('Suicide_Rates.csv')
# View Data
filtered_df.head(10)
```

```python
# Data Types
filtered_df.dtypes
```

```python
# change countries and year data types to string
filtered_df['country'] = filtered_df['country'].astype(str)
filtered_df['year'] = filtered_df['year'].astype(int)

filtered_df.dtypes
```

```python
filtered_df = filtered_df.drop("HDI for year", axis=1)
filtered_df = filtered_df.drop("generation", axis=1)
```

```python
filtered_df.dtypes
```

```python
# Remove 'years' suffixes by replacing with and empty string
filtered_df['age'] = filtered_df['age'].str.replace('years', '')
```

```python
# Filter 1990-2020
filtered_df = filtered_df.loc[(filtered_df['year'] >= 1990) & (filtered_df['year'] <= 2020)]
filtered_df.head()
print(filtered_df)
```

```python
filtered_df = filtered_df.loc[(filtered_df['country'] == 'United States') |  (filtered_df['country'] == 'France') |  (filtered_df['country'] == 'Mexico') |  (filtered_df['country'] == 'Germany') | (filtered_df['country'] == 'Brazil') |  (filtered_df['country'] == 'Canada') |  (filtered_df['country'] == 'Japan')]
filtered_df
```

```python
abreviation_df = filtered_df['Location'] = filtered_df['country'].str[:3]
#abreviation_df
filtered_df
```

```python
filtered_df['Location'] = ""
# Iterate through the rows of the DataFrame
#if len(row["country_code"]) != 3:
#       alpha_3 = country_name_to_country_alpha3(row["country"])
#        df.at[i, 'country_code'] = alpha_3

for i, row in filtered_df.iterrows():
    # Get the country name
    country_name = row['country']
    # Get the country object from pycountry library
    country = pycountry.countries.get(name=country_name)
    # Get the 3 letter alpha_3 code
    Location = country.alpha_3
    # Update the abbreviation column for the current row
    filtered_df.at[i, 'Location'] = Location
```
