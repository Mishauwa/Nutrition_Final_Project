```python
#Import libraries and read data csv file 
import pandas as pd
import os
```

```python
edu_data = os.path.join('Education_1990_2020.csv')
education_df = pd.read_csv(edu_data)
display(education_df)
```

```python
# Check the data types
education_df.dtypes
```

```python
# Converting Time int to string
education_df["TIME"]= education_df["TIME"].astype(str)
education_df
```

```python
#Combine Location and Time into Single Column
education_df["LocationTime"] = education_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame = education_df
education_df.head(10)
```

```python
# Change the Time Column dataformat back to int
education_df["TIME"]=education_df["TIME"].astype(int)
education_df.dtypes
```

```python
# Drop null values
education_df.isnull().any()
```

```python
#Drop Columns
education_df = education_df.drop(columns=['Flag Codes','FREQUENCY','MEASURE','INDICATOR'])
education_df.head(10)
```

```python
# Total number of rows
num_rows = len(education_df)
print(num_rows)
```

```python
# Select countries: USA, CAN, DEU, JPN, FRA
#education_df_filtered = education_df.loc[(education_df['LOCATION'] == 'USA') |  (education_df['LOCATION'] == 'FRA') |  (education_df['LOCATION'] == 'DEU') | (education_df['LOCATION'] == 'CAN') |  (education_df['LOCATION'] == 'JPN')]
#education_df_filtered
```

```python
# Filter TIME values between 1990 and 2020
education_df= education_df.loc[(education_df['TIME'] >= 1990) & (education_df['TIME'] <= 2020)]
education_df.head(10)
```

```python
# Move LocationTime row to the beginning of the DataFrame
move_column = education_df.pop('LocationTime')
education_df.insert(0, 'LocationTime', move_column)
education_df.head(10)
```

```python
# Rename value column
education_df=education_df.rename(columns={"Value":"Edu_val"})
education_df.head(10)
```

```python
# Using pivot table to get the subject column values
#education_df.pivot_table(index='LocationTime', columns='SUBJECT')
```

```python
education_df=pd.pivot(education_df, index = ["LocationTime", "LOCATION", "TIME"], columns=["SUBJECT"], values=["Edu_Val"])
education_df
```

```python
#format column names
education_df.columns = ['_'.join(str(s).strip() for s in col if s) for col in education_df.columns]
#reset index
education_df.reset_index(inplace=True)
education_df
```

```python
# Total number of rows
num_rows = len(education_df)
print(num_rows)
```

```python
# Save in a new csv file
education_df.to_csv('Education_data.csv', index=False)
```
