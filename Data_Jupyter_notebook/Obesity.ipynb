```python
import pandas as pd
```

```python
#import the data
#view top 10 
obesity_df= pd.read_csv("../Resources/Obesity_1990_2020.csv")
obesity_df.head(10)
```

```python
#change Datatype of time to Str
obesity_df["TIME"]=obesity_df["TIME"].astype(str)
```

```python
#Combine Location and Time into Single Column
obesity_df["LocationTime"] = obesity_df[["LOCATION", "TIME"]].apply("-".join, axis=1)
pd.DataFrame=obesity_df
```

```python
#change Datatype of time back to int
obesity_df["TIME"]=obesity_df["TIME"].astype(int)
```

```python
#filter to only include years 1990-2020
obesity_df_filter= obesity_df.loc[(obesity_df['TIME'] >= 1990) & (obesity_df['TIME'] <= 2020)]
obesity_df_filter=obesity_df_filter.drop(columns=["Flag Codes", "MEASURE", "FREQUENCY", "INDICATOR"]).rename(columns={"Value": "obesity_val"})
obesity_df_filter
```

```python
obesity_df=pd.pivot_table(obesity_df_filter, index = ["LocationTime", "LOCATION", "TIME"], columns=["SUBJECT"], values=["obesity_val"])
obesity_df
```

```python
#format column names
obesity_df.columns = ['_'.join(str(s).strip() for s in col if s) for col in obesity_df.columns]

#reset index
obesity_df.reset_index(inplace=True)
```

```python
obesity_clean=obesity_df.obesity_val_MEASURED.fillna(obesity_df.obesity_val_SELFREPORTED, inplace=True)
obesity_df.head()
```
