```python
# IMPORTING CSV FILE-
perc_data = os.path.join('New_Suicide_InValues.csv')
perc_df = pd.read_csv(perc_data)
display(perc_df)
```

```python
# percent = (len(list(filter(lambda ele: ele % 2 == 0, input_list))) / len(input_list)) * 100
perc_df['Alcohol_Val'] = perc_df['Alcohol_Val']*100
perc_df.head()
```

```python
perc_df['Healthy_Spend_Val'] = perc_df['Healthy_Spend_Val']*100
perc_df.head()
```

```python
p = perc_df.Suicide_Val_TOT
counts = p.value_counts()
counts
```

```python
percent = p.value_counts(normalize=True)
percent
```

```python
percent100 = p.value_counts(normalize=True).mul(100).round(1).astype(str) +'%'
percent100
```

```python
pd.DataFrame({'counts':counts, 'per':percent, 'per100':percent100})
```
